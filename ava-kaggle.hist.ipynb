```python
df = pd.read_csv('avocado.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head(5)
```

```python
df = df.drop(['4046','4225','4770','Large Bags','Small Bags','XLarge Bags','Total Volume'],axis=1)
```

```python
US = df.loc[(df['region']) == 'TotalUS']  
US.head(5)
```

```python
df['AveragePrice'].plot(kind='hist', rot=70, logx=True, logy=True)
# Season # 1= Spring #2= Summer #3= Fall #4= Winter
```

```python
# Create bee swarm plot with Seaborn's default settings
_ = sns.swarmplot(x='Season#', y='AveragePrice', data=US)

# Label the axes
_ = plt.xlabel('Season')
_ = plt.ylabel('Average Price')

# Show the plot
plt.show()
```
